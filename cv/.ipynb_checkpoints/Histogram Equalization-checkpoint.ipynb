```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import collections
```

```python
def plot_histogram(image):
  plt.figure(figsize=(8, 5), dpi=80)
  plt.hist(image.ravel(),256,[0,256])
  plt.show()
```

```python
def image_hist(img, normalized=True):
  m, n = img.shape
  h = np.zeros(256)
  
  for i in range(m):
    for j in range(n):
      h[img[i][j]] += 1
      
  if normalized:
    return np.array(h) / (m * n) 
  
  return np.array(h)
```

```python
def cumsum(hist):
  # finds cumulative sum of a numpy array
  return [sum(hist[:i+1]) for i in range(len(hist))]
```

```python
def hist_equal(img):
  
  his = image_hist(img)
  cdf = np.array(cumsum(his))
    
  tf = np.uint8(cdf * 255)
  
  m, n = img.shape
  new_img = np.zeros_like(img)
  
  for i in range(m):
    for j in range(n):
      new_img[i, j] = tf[img[i, j]]

  return tf, new_img
```

```python
# read original image and turn image in grayscale value
img = cv2.imread("test.jpg") # (194, 259)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plot_histogram(gray)
```

```python
# my plot histogram, normalized
hist = image_hist(gray, normalized=True)
plt.bar(range(0, 256), hist);
```

```python
# get transfer function and new image
tf, img_new = hist_equal(gray)
```

```python
plt.bar(range(0, 256), tf);
```

```python
plt.subplot(121)
plt.title("Old image")
plt.imshow(gray, cmap='gray')

plt.subplot(122)
plt.title("New image")
plt.imshow(img_new, cmap='gray');
```
